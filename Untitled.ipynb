```python
import pandas as pd
from automlbench import load_data, preprocess_data, get_models, evaluate_model, plot_performance, train_models


# Load dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Define target column
target_column = "Survived"

# Preprocess the data
X_train, X_test, y_train, y_test = preprocess_data(df, target_column)
```

```python
# Get default models
models = get_models()

# Train all models without tuning
results = train_models(X_train, X_test, y_train, y_test)

# Display results
print(results)
```

```python
# Evaluate models
for model_name, model in models.items():
    print(f"Evaluating {model_name}...")
    metrics = evaluate_model(model.fit(X_train, y_train), X_test, y_test)
    print(metrics)
```

```python
# Plot performance comparison
plot_performance(results)
```

```python
# Show performance comparison with all metrics
plot_performance(results, metrics=["Accuracy", "Precision", "Recall", "F1-Score", "RMSE"])
```
